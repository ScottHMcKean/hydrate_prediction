```python
import pandas as pd
import numpy as np
df = pd.read_parquet('/Users/scott.mckean/Desktop/3w_compiled_data.parquet')
df.sample(10)
```

```python
indices = ['timestamp', 'well_number']

vars = [
    'P-PDG',
    'P-TPT',
    'T-TPT',
    'P-MON-CKP',
    'T-JUS-CKP',
    'QGL'
    ]

target = ['state']
```

```python
x_df = (
    df[vars + indices + target]
    .sort_values(['well_number','timestamp'])
    .ffill().dropna()
    .copy()
)
```

```python
x_df.reset_index().to_parquet('/Users/scott.mckean/Desktop/3w_clean.parquet', coerce_timestamps="ms")
```
